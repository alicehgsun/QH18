### New json 
only made of title, url, time and lang(kor/en).

```python
import re
new_history = []
for elt in history:
    en_match = re.search(r"^[^가-힣ㄱ-ㅎㅏ-ㅣぁ-んァ-ン一-龥]*$", elt['title']) and len(elt['title'])>0
    kor_match = re.search(r"[ㄱ-ㅎㅏ-ㅣ가-힣]", elt['title']) and len(elt['title'])>0
    if en_match:
        new_history.append(dict(title = elt['title'], time = elt['lastVisitTime'], url = elt['url'], lang = 'en'))
    elif kor_match:
        new_history.append(dict(title = elt['title'], time = elt['lastVisitTime'], url = elt['url'], lang = 'kor'))
```

```python
len(new_history)
```

```python
with open('history2.json', 'w') as f:
     json.dump(new_history, f, ensure_ascii=False)
```
